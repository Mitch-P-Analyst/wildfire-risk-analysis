## Installs

```python
# Operational Packages
import numpy as np
import pandas as pd
from pathlib import Path
import sys
import os
import zipfile
import shutil

# Directories
nb_dir = Path.cwd()
REPO_ROOT = nb_dir.parent
data_dir = REPO_ROOT / 'data/'
processed_dir = data_dir / 'processed/'
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))
```

```python
def unzip_to_folder(zip_path, extract_to):
    """
    Unzips a ZIP archive into a specified directory.
    """
    extract_to = Path(extract_to)
    extract_to.mkdir(parents=True, exist_ok=True)

    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

    macosx_folder = extract_to / '__MACOSX'
    if macosx_folder.exists():
        shutil.rmtree(macosx_folder)
```

```python
zippaths = Path(data_dir/'raw/zips')
zipfolders = list(zippaths.glob('*.zip'))

for folder in zipfolders:
    unzip_to_folder(folder,processed_dir/str(folder.name)[:-4])

```
